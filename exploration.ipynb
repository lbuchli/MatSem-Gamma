```python
# euler limit definition of gamma

import matplotlib.pyplot as plt
import numpy as np

def f(x, N):
    """
    Computes the nth approximation of the Gamma function using Euler's limit definition:
    Gamma(z) = lim (n -> inf) (n! * n^z) / (z * (z+1) * ... * (z+n))
    """
    res = (N**x) / x
    for k in range(1, N + 1):
        res *= (k / (x + k))
    return res
    

x = np.linspace(-3, 6, 400)

threshold = 0.02
for pole in [0, -1, -2, -3, -4]:
    x = np.where(np.abs(x - pole) < threshold, np.nan, x)

plt.ylim(-5, 5)

ns = [1, 5, 10, 100]
colors = ['orange', 'red', 'green', 'blue']

for n, color in zip(ns, colors):
    y_approx = f(x, n)
    plt.plot(x, y_approx, label=f'Euler Limit ($n={n}$)', color=color)


plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
```
